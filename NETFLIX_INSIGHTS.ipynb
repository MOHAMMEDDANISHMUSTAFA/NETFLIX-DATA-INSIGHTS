```python
# Growth of Content over the release_year by Type of Show
content_by_type_year = df.groupby(['release_year', 'type'])['title'].count().unstack(fill_value=0)
content_by_type_year.plot(kind='line', marker='o', figsize=(20, 6))
plt.title('Growth of Content over the Years by Type of Show')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend(title='Type', loc='upper left')
plt.show()
```

```python
# Seasonality in Dates When Titles are Added
content_by_month = df.groupby('date_added')['title'].count()
plt.figure(figsize=(20, 10))
sns.barplot(x=content_by_month.index, y=content_by_month.values)
plt.title('Seasonality in Dates When Titles are Added')
plt.xlabel('Month')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
# Top 5 Movie Directors with the Longest Total Movie Duration
movie_directors_duration = df[df['type'] == 'Movie'].groupby('director')['duration'].sum()
top_directors = movie_directors_duration.sort_values(ascending=False).head(5)
print("Top 5 Movie Directors with the Longest Total Movie Duration:")
print(top_directors)
print()
```

```python
# Bonus: Top 5 Actors Seen on Netflix in the United States since 2018
top_actors_us = df[(df['type'] == 'Movie') & (df['release_year'] >= 2018)].groupby('cast')['title'].count()
top_actors_us = top_actors_us.sort_values(ascending=False).head(5)
print("Top 10 Actors Seen on Netflix in the United States since 2018:")
print(top_actors_us)
print()
```

```python
# Visualization 3: Pie Chart of Top 5 Actors Seen on Netflix in the United States since 2018
plt.figure(figsize=(8, 8))
plt.pie(top_actors_us, labels=top_actors_us.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Actors Seen on Netflix in the United States since 2018')
plt.axis('equal')
plt.show()
```
